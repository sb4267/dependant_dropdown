```python
import pandas as pd
df = pd.read_csv("new_autos_dropped.csv")
```

```python
brands=[]
for i in df['brand'].unique():
    brands.append(i)
```

```python
brands_df = pd.DataFrame(brands, columns = ['brand_name'])
brands_df
```

```python
brand_id=0
super_model=[]
for i in df['brand'].unique():
    data_model=[]
    for j in range(len(df['brand'])):
        if df['brand'][j]==i:
            data_model.append(df['model'][j])
    model=set(data_model)
    for a in model:
        super_model.append([a,brand_id])
    brand_id+=1
```

```python
super_model_df = pd.DataFrame(super_model, columns = ['brand_id','model_name_id'])
super_model_df
```

```python
indexing=0
for each in super_model:
    each.insert(0, indexing)
    indexing+=1
```

```python
super_model
```

```python
super_vehicle=[]
for x in super_model:
    model_data=[]
    for i in range(len(df['model'])): 
        if x[1] == df['model'][i]:
            model_data.append(df['vehicle_type'][i])
    model=set(model_data)
    for each in model:
        super_vehicle.append([each,x[0]])
```

```python
super_vehicle_df = pd.DataFrame(super_vehicle, columns = ['name','model_n_id'])

super_vehicle_df
```

```python
import sqlite3
conn = sqlite3.connect("db.sqlite3")
```

```python
conn.close
```

```python
brands_df.to_sql("v3_brand", conn, if_exists="replace", index_label='id')
```

```python
super_model_df.to_sql("v3_models_name", conn, if_exists="replace",index_label='id')
```

```python
super_vehicle_df.to_sql("v3_v_type", conn, if_exists="replace",index_label='id')
```
